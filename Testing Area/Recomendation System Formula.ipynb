```python
import pandas as pd
def recomendation_engine(product,SKU,UserID,ratings,max_score = 4):
    working_df = pd.DataFrame(dict(SKU = SKU, UserID = UserID, ratings=ratings)).reset_index(drop=True)
    ratings_mtx_df = working_df.pivot_table(index='UserID', columns='SKU',values='ratings')
    ratings_gte_4 = ratings_mtx_df[ratings_mtx_df>=max_score]
    ratings_gte_4_pd = pd.DataFrame({'total': ratings_mtx_df.count(), 'gte_4': ratings_gte_4.count()})
    ratings_gte_4_pd['gte_4_ratio'] = (ratings_gte_4_pd['gte_4'] * 1.0)/ ratings_gte_4_pd.total
    ranking = [(title,total,gte_4, score) for title, total, gte_4, score in ratings_gte_4_pd.itertuples()]
    ratings_by_title = working_df.groupby('SKU').size()
    product_users = working_df.pivot_table(index='SKU', columns='UserID',values='ratings')
    ratings_by_product = working_df[working_df['SKU'] == product]
    ratings_by_product.set_index('UserID', inplace=True)
    their_ids = ratings_by_product.index
    their_ratings = product_users[their_ids]
    product_count =  their_ratings.loc[product].count()
    sims = their_ratings.apply(lambda profile: profile.count() / float(product_count) , axis=1)
    return sims.sort_values(ascending=False)[1:][:10]
   
    
```
