```python
Trans_per_user = sku_df.merge(transID_df, on='ID')
trans_with_name = Trans_per_user.merge(product_df, on='SKU')
trans_with_name.reset_index()
print(len(trans_with_name))
trans_with_name.head()
```

```python
final_db = trans_with_name.dropna()
print(len(final_db))
final_db.head()
```

```python
# Creating ratings based on ideas
```

```python
import numpy as np
from numpy import float32
import pandas as pd

a=.7 #3
b=1-a

final_db['price'] = final_db['Transaction_Revenue'].div(final_db['qty'])
print(len(final_db))
Db_sans_1 = final_db[final_db['price']>1]

Db_sans_1['ratings'] = np.log(Db_sans_1['qty'])*a + np.log(Db_sans_1['price'])*b
Db_sans_1.reset_index(inplace=True,drop=True)

Db_sans_1.sort_values(by='ratings',ascending=False).head()
```

```python
import datetime
trash_out = r'C:\Users\Gpardo\Desktop\proyecto_test\Final_db'
save = datetime.datetime.now().strftime("%y_%m_%d")
dir = trash_out +"_"+str(save)+".xlsx"
print(dir)

writer = pd.ExcelWriter(dir, engine='xlsxwriter')

#Convert the dataframe to an XlsxWriter Excel object.
Db_sans_1.to_excel(writer, sheet_name='Sheet1')

# Close the Pandas Excel writer and output the Excel file.
writer.save()
```
